```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world-happiness-report-2021.csv')
```

```python
df.info()
```

```python
cols = list(df.columns)
new_cols = (cols[0:3])
new_cols.extend(cols[6:12])
new_cols
```

```python
df = df[new_cols]
```

```python
sns.heatmap(df.corr())
```

```python
hle = df['Healthy life expectancy']
lgdp = df['Logged GDP per capita']
```

```python
#plt.subplots(figsize=(10, 8)) scatter_kws={"color": "black"}, line_kws={"color": "red"}
sns.lmplot(x="Healthy life expectancy", y="Logged GDP per capita", data=df, hue='Regional indicator', palette="Set1", aspect=1.5)

print("Correlation coefficient: ", np.corrcoef(hle, lgdp), sep="\n")
print("Spearman's rho for Healthy life expectancy: %.2f" % hle.corr(lgdp,method='spearman'))
print("Kendall's tau for Healthy life expectancy: %.2f" % hle.corr(lgdp,method='kendall'))

# this code correlates the Healthy life expectancy and the Logged GDP per capita using linear regression per Region. Furthermore, the correlation coefficient, # Spearman's rho and Kendall's tau for all data (not categorizing) is estimated
```
