```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SCP_2020-21.csv")
```

```python
a = pd.melt(df, id_vars=["Player's Name"], var_name = "Goals in a match", value_name="Goals")
a.groupby(["Player's Name"]).describe()
```

```python
plt.figure(figsize=(20,5))
g = sns.boxplot(x="Player's Name", y="value", data=pd.melt(df, id_vars=["Player's Name"]))
g.set_xticklabels(g.get_xticklabels(), rotation=40, ha="right")

plt.xlabel("Player's Names")
plt.ylabel("Goals per match")
plt.title("SCP players' stats: goals per match")
plt.show()
```
